```python
book_count_df = df.groupby(df['genre1']).size().sort_values().reset_index()
book_count_df.rename(columns={0:'count'}, inplace=True)
book_count_df
```

```python
fig = px.bar(book_count_df, x='genre1', y='count',
              title = '2001 ~ 2021 Aladin book bount in rank_100',
              color='genre1',
              hover_name='genre1')

fig.show()
```

```python
fig = px.pie(book_count_df, values='count', names='genre1',
             title='2001 ~ 2021 Aladin book count in rank_100',
             hover_data=['count'], labels={'count':'genre1'})

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
df20 = df[df['rank'] <= 20]
df20.tail()
```

```python
df20.groupby(['year', 'genre1'])['rank'].min().reset_index()
```

```python
portion_cnt = df.groupby(['year','genre1']).size().reset_index()
portion_cnt.rename(columns={0:'count'}, inplace=True)
portion_cnt
```

```python
fig = px.line(portion_cnt, x='year', y='count',
              title = '2001 ~ 2021 Aladin book_count in rank_100',
              color='genre1',
              line_group='genre1',
              hover_name='genre1')

fig.show()
```

```python
portion20_cnt = df20.groupby(['year','genre1']).size().reset_index()
portion20_cnt.rename(columns={0:'count'}, inplace=True)
portion20_cnt
```

```python
fig = px.line(portion20_cnt, x='year', y='count',
              title = '2001 ~ 2021 Aladin book count in rank_20',
              color='genre1',
              line_group='genre1',
              hover_name='genre1')

fig.show()
```

```python
max_rank = df20.groupby(['year', 'genre1'])['rank'].min().reset_index()
max_rank
```

```python
fig = px.line(max_rank, x='year', y='rank',
              title = '2001 ~ 2021 Aladin book top rank in rank_20',
              color='genre1',
              line_group='genre1',
              hover_name='genre1')
fig.update_yaxes(range=[21, 0])
fig.update_xaxes(range=[1999, 2022])

fig.show()
```

```python
social = pd.read_csv('2001_2021_798_7월_순위.csv')
```

```python
social.head()
```

```python
social_cnt = social.groupby(['genre2']).size().sort_values().reset_index()
social_cnt.rename(columns={0:'count'}, inplace=True)
social_cnt = social_cnt[social_cnt['count'] >=7 ]

fig = px.bar(social_cnt, x='genre2', y='count',
              title = '2001 ~ 2021 Aladin book bount in social rank_100',
              color='genre2',
              hover_name='genre2')

fig.show()
```

```python
social10 = social[social['rank'] <= 10]
social10_cnt = social10.groupby(['genre2','year']).size().sort_values().reset_index()
social10_cnt.rename(columns={0:'count'}, inplace=True)
social10_cnt
```

```python
fig = px.bar(social10_cnt, x = 'year', y = 'count', color = 
    'genre2', barmode = 'stack')

fig.update_layout(title = "2001 ~ 2021 book count in social top 10",
     xaxis_title = 'year', yaxis_title = 'count')
fig.show()
```

```python
economics = pd.read_csv('2001_2021_170_7월_순위.csv')
```

```python
economics.head()
```

```python
economics10 = economics[economics['rank'] <= 10]
economics10_cnt = economics10.groupby(['genre2','year']).size().sort_values().reset_index()
economics10_cnt.rename(columns={0:'count'}, inplace=True)
economics10_cnt
```

```python
fig = px.bar(economics10_cnt, x = 'year', y = 'count', color = 
    'genre2', barmode = 'stack')

fig.update_layout(title = "2001 ~ 2021 book count in economics top 10",
     xaxis_title = 'year', yaxis_title = 'count')
fig.show()
```
